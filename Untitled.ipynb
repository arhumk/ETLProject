```python
from splinter import Browser
from bs4 import BeautifulSoup
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.yelp.com/search?find_desc=&find_loc=East+Harlem%2C+Manhattan%2C+NY&ns=1'
browser.visit(url)
```

```python
price_ranges=[]
for x in range(1, 12):

    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')

    price_levels = soup.\
        find_all('span',class_="lemon--span__373c0__3997G text__373c0__3s1FV priceRange__373c0__2DY87 text-color--normal__373c0__VIFNO text-align--left__373c0__fEFN5 text-bullet--after__373c0__3YzzC")

    
    for price_level in price_levels:
        print('page:', x, '-------------')
        print(price_level.text)
        price_ranges.append(price_level.text)

    
    #browser.find_by_text("Next").first.click()
    try:
        browser.click_link_by_partial_text('Next')
          
    except:
        print("Scraping Paused")
    
    
```
