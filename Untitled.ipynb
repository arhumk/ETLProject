```python
from splinter import Browser
from bs4 import BeautifulSoup
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.yelp.com/search?find_desc=&find_loc=east+harlem&ns=1'
browser.visit(url)
```

```python
def click_next(browser, tryNo=0):
    if tryNo > 10:
        raise Exception("Nope")
    try:
        browser.find_by_text("Next").last.click()
    except:
        click_next(browser, tryNo=tryNo+1)
```

```python
price_ranges=[]

for x in range(1, 12):

    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')

    price_levels = soup.\
        find_all('span',class_="lemon--span__373c0__3997G text__373c0__2pB8f priceRange__373c0__2DY87 text-color--normal__373c0__K_MKN text-align--left__373c0__2pnx_ text-bullet--after__373c0__1ZHaA")
    
    
    
    for price_level in price_levels:
        print('page:', x, '-------------')
        print(price_level.text)
        price_ranges.append(price_level.text)
        
    
    
    
    browser.is_text_present('Next', wait_time=10)
    click_next(browser)
    
    
    
```
